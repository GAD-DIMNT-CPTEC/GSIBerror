```python
# Uncomment next line to use in Google Colab
#!pip install GSIBerror

import gsiberror as gb

import os

%matplotlib inline
```

```python
# If you are in Google Colab, use:
#!wget http://ftp1.cptec.inpe.br/pesquisa/das/carlos.bastarz/GSIBerror/data/bcptec_bam_352pairs_gsir4.berror_stats.fix.oz.cw.tsm.gcv
#!wget http://ftp1.cptec.inpe.br/pesquisa/das/carlos.bastarz/GSIBerror/data/global_berror.l64y386.f77-ncep-dtc.gcv
#path = './'
# If you are running Jupyter locally, use:
path = '../data'

filenames = {
    'file0' :['global_berror.l64y386.f77-ncep-dtc.gcv',                   'NCEP_GFS_hyb'],
    'file1' :['new_gsir4.berror_stats-tese-cfbastarz2017.gcv',            'CPTEC_MCGA_sig-ref'],
    'file2' :['bcptec_bam_352pairs_gsir4.berror_stats.fix.oz.cw.tsm.gcv', 'CPTEC_BAM_hyb-ref']
}
```

```python
matrix = []

for file in filenames.values():
    tmp1 = os.path.join(path,file[0]) # join path and file name
    tmp2 = gb.Berror(tmp1)               # instantiate a matrix
    tmp2.my_name(file[1])             # name a matrix
    tmp2.read_records()               # read the records
    matrix.append(tmp2)               # append the matrix a list of matrices
```

```python
Vars = ['agvin', 'bgvin', 'wgvin']

lmatrix = [matrix[0], matrix[2]]

for var in Vars:
    gb.plot_reg_coeffs(lmatrix, var, 0, eqrange=False, suptitle=True, savefig=False) # 0 = primeiro nível
```

```python
# Note

# The records of ozone, cloud water and sea surface temperature (amplitudes and horizontal/vertical length 
# scales) are intentionally the same for both matrices displayed in this notebook. In fact, the CPTEC **B** 
# matrix borrows these records from the NCEP **B** matrix (provides by the Developmental Testbed Center_ - 
# DTC) for testing purposes.

Vars = ['sf', 'vp', 't', 'q', 'qin', 'oz', 'ps', 'cw', 'sst']

lmatrix = [matrix[0], matrix[2]]

for var in Vars:
    gb.plot_amplitudes(lmatrix, var, profile=False, eqrange=False)
    if var == 'ps' or var == 'sst': # skip the vertical profile plotting for the ps and sst
        continue
    else:
        gb.plot_amplitudes(lmatrix, var, profile=True, eqrange=False, suptitle=True, savefig=False)
```

```python
Vars = ['sf', 'vp', 't', 'q', 'oz', 'cw', 'ps', 'sst']

lmatrix = [matrix[0], matrix[2]]

for var in Vars:
    gb.plot_hscales(lmatrix, var, eqrange=False, suptitle=True, savefig=False)
```

```python
Vars = ['sf', 'vp', 't', 'q', 'oz', 'cw']

lmatrix = [matrix[0], matrix[2]]

for var in Vars:
    gb.plot_vscales(lmatrix, var, eqrange=False, suptitle=True, savefig=False)
```
