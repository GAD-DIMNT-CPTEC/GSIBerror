```python
from GSIBerror import Berror

import os
import plot_functions as pf

%matplotlib inline
```

```python
path = 'gcv/'

filenames = {
    'file0' :['global_berror.l64y386.f77-ncep-dtc.gcv',                   'NCEP_GFS_hyb'],
    'file1' :['new_gsir4.berror_stats-tese-cfbastarz2017.gcv',            'CPTEC_MCGA_sig-ref'],
    'file2' :['bcptec_bam_352pairs_gsir4.berror_stats.gcv.fix.oz.cw.tsm', 'CPTEC_BAM_hyb-ref']
}
```

```python
matrix = []

for file in filenames.values():
    tmp1 = os.path.join(path,file[0]) # join path and file name
    tmp2 = Berror(tmp1)               # instantiate a matrix
    tmp2.my_name(file[1])             # name a matrix
    tmp2.read_records()               # read the records
    matrix.append(tmp2)               # append the matrix a list of matrices
```

```python
Vars = ['agvin', 'bgvin', 'wgvin']

lmatrix = [matrix[0], matrix[2]]

for var in Vars:
    pf.plot_reg_coeffs(lmatrix, var, 0, eqrange=False, suptitle=True, savefig=False) # 0 = primeiro nível
```

```python
Vars = ['sf', 'vp', 't', 'q', 'qin', 'oz', 'ps', 'cw', 'sst']

lmatrix = [matrix[0], matrix[2]]

for var in Vars:
    pf.plot_amplitudes(lmatrix, var, profile=False, eqrange=False)
    if var == 'ps' or var == 'sst': # não plota os perfis para ps e sst
        continue
    else:
        pf.plot_amplitudes(lmatrix, var, profile=True, eqrange=False, suptitle=True, savefig=False)
```

```python
Vars = ['sf', 'vp', 't', 'q', 'oz', 'cw', 'ps', 'sst']

lmatrix = [matrix[0], matrix[2]]

for var in Vars:
    pf.plot_hscales(lmatrix, var, eqrange=False, suptitle=True, savefig=False)
```

```python
Vars = ['sf', 'vp', 't', 'q', 'oz', 'cw']

lmatrix = [matrix[0], matrix[2]]

for var in Vars:
    pf.plot_vscales(lmatrix, var, eqrange=False, suptitle=True, savefig=True)
```
